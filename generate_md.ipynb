```python
import json
import urllib.parse

base_url = 'https://cloud.mail.ru/public/'
genre_string = '''[Назад](../README.md)

# {}
'''
game_string = '''<details>
  <summary>{}</summary>

| Жанр | {} |
| - | - |
| URL | {} |

> {}
</details>

'''
core_md_genre_string = '''### [Каталог: {}](generated/out-{}.md)  '''

out_all_games_string_header = '''[Назад](../README.md)

# Все игры списком  
| Имя | URL |
| - | - |'''

out_all_games_string = '''
| {} | {} |'''

output_data = {}

with open('base_games.json') as f:
    data = json.load(f)

    for item in data:
        game = {}
        game['title'] = item['title']
        game['url'] = item['url']
        game['setup_folder'] = item['games_folder']
        game['genre_fold'] = item['genre_fold']
        game['genre'] = item['genre']
        game['note'] = item['note']

        output_data.setdefault(item['genre_fold'], []).append(game)

for genre in output_data:
    with open("generated/out-{}.md".format(genre), "w", encoding="utf-8") as out_genre:
        out_genre.write(genre_string.format(genre))
        
        for game in output_data[genre]:
            url = '{}{}/{}'.format(base_url, game['url'], urllib.parse.quote(game['setup_folder']))
            
            game_data = game_string.format(
                game['title'],
                game['genre'],
                url,
                game['note']
            )
            out_genre.write(game_data)

with open("generated/out-all.md", "w", encoding="utf-8") as out_all:
    out_all.write(out_all_games_string_header)
    
    for genre in output_data:
        for game in output_data[genre]:
            url = '{}{}/{}'.format(base_url, game['url'], urllib.parse.quote(game['setup_folder']))
            
            game_data = out_all_games_string.format(
                game['title'], url,
            )
            out_all.write(game_data)
            print(game['title'])

```
